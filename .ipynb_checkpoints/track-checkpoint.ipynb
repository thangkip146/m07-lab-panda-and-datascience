```python
import pandas as pd

df = pd.read_csv(
    "track.csv",
    parse_dates=['time'],
    index_col='time'
)
df = df.resample('min').mean()
```

```python
import folium

Map = folium.Map(
    location = [df['lat'].mean(), df['lng'].mean()],
    zoom_start=15,
)

def add_marker(row):
    loc = tuple(row[['lat','lng']])
    marker = folium.CircleMarker(
        loc,
        radius=5,
        color='blue' if row['height'] < 100 else 'red',
        popup=row.name.strftime('%H:%M'),
    )
    marker.add_to(Map)
    
df.apply(add_marker, axis=1)
Map
```
